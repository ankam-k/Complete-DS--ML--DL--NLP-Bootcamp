### Assignment 6: List Methods

Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

```python
import random

lst=[random.randint(1, 10) for i in range(5)]
print(lst)
# Sort the list in ascending order
sort_asc=sorted(lst) 
print(sort_asc)
# Sort the list in descending order
sort_dsc=sorted(lst, reverse=True) 
print(sort_dsc)
# Remove duplicates
unique_l=list(set(lst))
print(unique_l)
```

### Assignment 8: List of Dictionaries

Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

```python
students=[
    {"name": "ABC", "score": 85},
    {"name": "DEF", "score": 92},
    {"name": "GHI", "score": 78}
]

# Sort students by score in descending order
sort_students=sorted(students, reverse=True, key=lambda x:x['score'])
for i in sort_students:
    print(f"Name: {i['name']}, Score: {i['score']}")
```

### Assignment 9: Matrix Transposition

Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

```python
def transpose_matrix(matrix):
    transpose_mat=[[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
    return transpose_mat


matrix=[[1,2,3],[4,5,6]]
transpose_mat= transpose_matrix(matrix)

for i in range(len(transpose_mat)):
    for j in range(len(transpose_mat[0])):
        print(transpose_mat[i][j], end=" ")
    print() 
```

### Assignment 10: Flattening a Nested List

Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

```python
def flatten_list(nested_list):
    flat_list=[i for sublist in nested_list for i in sublist]
    return flat_list

nested_list=[[1,3],[2,4],[5,7],[6,8]]
print("Nested List:", nested_list)
flat_list=flatten_list(nested_list)
print("Flattened List:", flat_list)
flat_list.sort()
print("Sorted Flattened List:", flat_list)
sorted_dsc=sorted(flat_list, reverse=True)
print("Sorted Flattened List (Descending):", sorted_dsc)
print("List Elements(Matrix form): ")
for i in nested_list:
    for j in i:
        print(j, end=" ")
    print()
print("List Elements(Transposed matrix form): ")
transpose_list=[[nested_list[j][i] for j in range(len(nested_list))] for i in range(len(nested_list[0]))]
for i in transpose_list:
    for j in i:
        print(j, end=" ")
    print()
```

### Assignment 13: List Reversal

Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.

```python
def reverse_list(lst):
    return lst[::-1]

lst=[1,9,2,8,3,7,4,6,5]
print(lst)
rev_list=reverse_list(lst)
print(rev_list)
```

### Assignment 14: List Rotation

Write a function that rotates a list by n positions. Print the original and rotated lists.

```python
def rotate_list(lst, n):
    return lst[n:] + lst[:n]

original_list = [1, 2, 3, 4, 5]
rotated_list = rotate_list(original_list, 2)
print(f"Original list: {original_list}")
print(f"Rotated list: {rotated_list}")
```

```python
def rotate_list_left(lst, n):
    return lst[n:]+lst[:n]

def rotate_lst_right(lst, n):
    return lst[-n:]+lst[:-n]

lst11=[1,2,3,4,5]
print("Original list", lst11)
rotated_lst=rotate_list_left(lst11, 3)
print("Rotated list (left)", rotated_lst)
rotated_right=rotate_lst_right(lst11, 3)
print("Rotated list (right)", rotated_right)
```

### Assignment 15: List Intersection

Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

```python
def intersect(lst1,lst2):
    intersection_list=[i for i in lst1 if i in lst2]
    return intersection_list

lst10=[1,2,3,4,5]
list10=[1,3,5,7,9]
inter=intersect(lst10, list10)
print("Intersection of lst10 and list10:", inter)
```
